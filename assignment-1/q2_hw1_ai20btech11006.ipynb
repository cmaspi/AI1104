```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python

#number of points in each cluster initially
num_pts=int(2e3)

#data dimension p, and number of clusters k
p=int(input("Enter the value of data dimension: "))
k=int(input("Enter the number of clusters: "))

#covariance matrix
cov=np.identity(p, dtype=float)

#mean matrix, k*p dimensional
mean=[]

# mean matrix contains random integer from -2k^2 to 2k^2 with step size 2
for i in range(k):
    row=[]
    for j in range(p):
        row.append(random.randrange(-2*k*p,2*k*p))
    mean.append(row)
    print(mean)


# generating the points in cluster
Y=np.random.multivariate_normal(mean[0],cov,num_pts).T
for i in range(k-1):
    y=np.random.multivariate_normal(mean[i+1],cov,num_pts).T
    Y=np.concatenate((Y,y),axis=1)



```

```python
#tolerance in error
epsilon=0.001

#error, initialised with large value
error=1e6

#generating random centroids to begin with
D=np.random.multivariate_normal(Y.mean(1),200*np.identity(p,dtype=float),k).T

```

```python
# it represents the iteration
count=0

#this array represents which datapoint is assigned to which centroid
cluster=[]
# epsilon which is the tolerance should be lower than error
while(error>epsilon):
    cluster.clear()
    num=[0]*k
    # X is k*p matrix
    X=[[0]*p for i in range(k)]
    
    for idx in range(k*num_pts):
        cur_y=Y[:,idx]
        d=[0]*k
        
        #computing the distance
        for i in range(k):
            for j in range(p):
                d[i]+=np.square(cur_y[j]-D[j,i])
            d[i]=np.sqrt(d[i])

        min_idx=d.index(np.min(d))
        for i in range(p):
            X[min_idx][i]+=cur_y[i]
        num[min_idx]+=1
        
        #assigning a centroid to a datapoint
        cluster.append(min_idx)
    
    #calculating the centroids
    for i in range(k):
        for j in range(p):
            if num[i]==0:
                break
            X[i][j]/=num[i]
    error=0
    #computing the error
    for i in range(k):
        for j in range(p):
            if num[i]==0:
                break
            error+=np.square(X[i][j]-D[j][i])
    error=np.sqrt(error)

    #update the centroids
    for i in range(k):
        for j in range(p):
            if num[i]==0:
                break
            D[j][i]=X[i][j]
    #update count
    count+=1


        # Print error
    print ('Iteration: ', count, 'Error: ', error)

# Print centroids
print ('Centroids:', D)

# Final assignment of data point to clusters
print('cluster assigned to data point',cluster)
```
